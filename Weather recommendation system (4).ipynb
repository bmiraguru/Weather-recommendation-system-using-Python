```python
import numpy as np
import statsmodels.api as sm
import scipy.stats as stats

# Extracting columns for analysis
X = df['Temp']
y = df['Feels Like']

# Adding a constant to the independent variable (for intercept)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print model summary
print(model.summary())

# Make predictions
y_pred = model.predict(X)

# Residuals
residuals = y - y_pred

# Scatter plot of observed vs. predicted
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted')
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Load the data
data = pd.DataFrame({
    'Temp': [45, 40, 35, 30, 40, 30, 25, 40, 40, 40, 40, 20, 40, 45, 40, 30, 35, 40, 35, 50, 40, 45, 40, 40, 50, 40, 40],
    'Feels Like': [35, 30, 30, 20, 35, 25, 15, 35, 40, 35, 35, 10, 40, 45, 40, 20, 30, 30, 30, 45, 35, 40, 40, 35, 45, 35, 35]
})

# Fit the linear regression model
X = sm.add_constant(data['Temp'])  # Add a constant to the independent variable
model = sm.OLS(data['Feels Like'], X).fit()

# Get residuals from the model
residuals = model.resid

# Create a histogram of residuals
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True, color='skyblue')
plt.title('Histogram of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv("Running Weather - Log.csv")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# Assuming your DataFrame contains 'Feels Like' and 'Temp' columns
actual_temperatures = df['Temp'].values.reshape(-1, 1)
feels_like_temperatures = df['Feels Like'].values

# Train a linear regression model
model = LinearRegression()
model.fit(actual_temperatures, feels_like_temperatures)

# Function to predict feels like temperature based on actual temperature
def predict_feels_like_temperature(actual_temperature):
    predicted_feels_like_temperature = model.predict([[actual_temperature]])
    return predicted_feels_like_temperature[0]

# User input for actual temperature
actual_temperature = float(input("Enter the actual temperature: "))

# Predict feels like temperature based on actual temperature
predicted_feels_like_temperature = predict_feels_like_temperature(actual_temperature)
print("Feels like temperature:", predicted_feels_like_temperature)

# Make predictions
predicted_feels_like_temperatures = model.predict(actual_temperatures)

# Calculate model accuracy metrics
mae = mean_absolute_error(feels_like_temperatures, predicted_feels_like_temperatures)
mse = mean_squared_error(feels_like_temperatures, predicted_feels_like_temperatures)
r2 = r2_score(feels_like_temperatures, predicted_feels_like_temperatures)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)
```

```python
import pandas as pd

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract year and month from the date
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Group by year and month, then calculate the monthly average temperature
monthly_avg_temp = df.groupby(['Year', 'Month'])['Temp'].mean().reset_index()

# Display the monthly average temperatures
print(monthly_avg_temp)


# Plotting the data
plt.figure(figsize=(10, 6))  # Set the figure size (width, height)

# Extract years for plotting (assuming you have data spanning multiple years)
years = monthly_avg_temp['Year'].unique()

for year in years:
    year_data = monthly_avg_temp[monthly_avg_temp['Year'] == year]
    plt.plot(year_data['Month'], year_data['Temp'], label=str(year))

# Customize the plot
plt.title('Monthly Average Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year', loc='upper right')
plt.grid(False)

# Show the plot
plt.show()
```
